```python
import sqlite3, re
import pandas as pd

def regexp(expr, item):
    reg = re.compile(expr)
    return reg.search(item) is not None

connect = sqlite3.connect('download_stats.db')
connect.create_function("REGEXP", 2, regexp)
# keyword in SQL command; # of args of self-defined function; self-defined function 
cursor = connect.cursor()
cursor.execute('select * from download where source REGEXP(?)', ['210906'])
cursor.fetchall()
```

```python
import sqlite3, re
import pandas as pd

def regexp(expression, item):
    return re.compile(expression).search(item) != None

connect = sqlite3.connect('download_stats.db')
connect.create_function("REGEXP", 2, regexp)
pd.read_sql("select * from download where source REGEXP(?)", con=connect, \
    params=['21.906'])
```
